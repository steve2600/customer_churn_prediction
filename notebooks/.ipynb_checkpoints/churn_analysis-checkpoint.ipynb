```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


pd.set_option('display.max_columns', None)


print(" Libraries loaded successfully!")
```

```python
# Cell 2: Load Data
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# First look at data
print(f"Dataset shape: {df.shape}")
print(f"We have {df.shape[0]} customers and {df.shape[1]} features")
print("\nFirst 5 rows:")
df.head()
```

```python
# Cell 3: Data Overview
print("="*60)
print("COLUMN NAMES")
print("="*60)
for i, col in enumerate(df.columns, 1):
    print(f"{i}. {col}")

print("\n" + "="*60)
print("DATA TYPES")
print("="*60)
print(df.dtypes)

print("\n" + "="*60)
print("MISSING VALUES")
print("="*60)
missing = df.isnull().sum()
if missing.sum() == 0:
    print("No missing values!")
else:
    print(missing[missing > 0])

print("\n" + "="*60)
print("CHURN DISTRIBUTION")
print("="*60)
print(df['Churn'].value_counts())
print(f"\nChurn Rate: {(df['Churn']=='Yes').sum() / len(df) * 100:.2f}%")
```

```python
# Cell 4: Fix TotalCharges column
# This column has spaces instead of numbers for some customers

# Check current data type
print(f"TotalCharges type: {df['TotalCharges'].dtype}")

# Find problem rows
problem_rows = df[df['TotalCharges'] == ' ']
print(f"\nFound {len(problem_rows)} rows with spaces")

# Convert to numeric (spaces become NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Check how many NaN
print(f"NaN values: {df['TotalCharges'].isna().sum()}")

# Fill NaN with 0 (these are new customers with 0 tenure)
df['TotalCharges'].fillna(0, inplace=True)

print("\n TotalCharges fixed!")
```

```python

print("="*60)
print("INSIGHT 1: TENURE ANALYSIS")
print("="*60)
tenure_stats = df.groupby('Churn')['tenure'].agg(['mean', 'median', 'min', 'max'])
print(tenure_stats)
plt.figure(figsize=(10, 6))

df[df['Churn'] == 'No']['tenure'].hist(
    bins=30, alpha=0.6, label='Stayed (No Churn)', color='green', edgecolor='black'
)
df[df['Churn'] == 'Yes']['tenure'].hist(
    bins=30, alpha=0.6, label='Left (Churned)', color='red', edgecolor='black'
)
plt.xlabel('Tenure (months)', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.title('Customer Tenure Distribution by Churn Status', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../images/01_tenure_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n Plot saved to images/01_tenure_distribution.png")
```

```python

print("="*60)
print("INSIGHT 2: CONTRACT TYPE ANALYSIS")
print("="*60)

contract_churn = df.groupby('Contract')['Churn'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).round(2)

print("Churn Rate by Contract Type:")
print(contract_churn)
print(f"\nOverall churn rate: {(df['Churn']=='Yes').sum() / len(df) * 100:.2f}%")

plt.figure(figsize=(10, 6))
bars = plt.bar(contract_churn.index, contract_churn.values, 
               color=['#e74c3c', '#f39c12', '#2ecc71'], 
               edgecolor='black', linewidth=1.5)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%',
             ha='center', va='bottom', fontsize=12, fontweight='bold')

avg_churn = (df['Churn']=='Yes').sum() / len(df) * 100
plt.axhline(y=avg_churn, color='red', linestyle='--', linewidth=2, 
            label=f'Average: {avg_churn:.1f}%')

plt.xlabel('Contract Type', fontsize=12, fontweight='bold')
plt.ylabel('Churn Rate (%)', fontsize=12, fontweight='bold')
plt.title('Churn Rate by Contract Type', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../images/02_contract_churn.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nPlot saved to images/02_contract_churn.png")
```

```python

print("="*60)
print("INSIGHT 3: MONTHLY CHARGES ANALYSIS")
print("="*60)

charges_stats = df.groupby('Churn')['MonthlyCharges'].describe()
print(charges_stats)

# Visualization - Boxplot
plt.figure(figsize=(10, 6))
bp = df.boxplot(column='MonthlyCharges', by='Churn', 
                patch_artist=True, return_type='dict')

colors = ['lightgreen', 'lightcoral']
for patch, color in zip(bp['MonthlyCharges']['boxes'], colors):
    patch.set_facecolor(color)

plt.suptitle('')  
plt.title('Monthly Charges Distribution by Churn Status', 
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Churn Status', fontsize=12, fontweight='bold')
plt.ylabel('Monthly Charges ($)', fontsize=12, fontweight='bold')

for i, churn_val in enumerate(['No', 'Yes']):
    mean_val = df[df['Churn']==churn_val]['MonthlyCharges'].mean()
    plt.text(i+1, mean_val, f'Mean: ${mean_val:.2f}', 
             ha='center', va='bottom', fontsize=10, 
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../images/03_monthly_charges.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nPlot saved to images/03_monthly_charges.png")
```

```python
# Cell 8: Correlation Analysis

print("="*60)
print("CORRELATION ANALYSIS")
print("="*60)
df_corr = df.copy()
df_corr['Churn_Numeric'] = (df_corr['Churn'] == 'Yes').astype(int)
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn_Numeric']

correlation = df_corr[numeric_cols].corr()['Churn_Numeric'].sort_values(ascending=False)
print(correlation)

plt.figure(figsize=(10, 8))
correlation_matrix = df_corr[numeric_cols].corr()

sns.heatmap(correlation_matrix, annot=True, fmt='.3f', 
            cmap='coolwarm', center=0, square=True,
            linewidths=1, cbar_kws={"shrink": 0.8})

plt.title('Correlation Matrix', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('../images/04_correlation.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nPlot saved to images/04_correlation.png")
```

```python

print("="*60)
print("FEATURE ENGINEERING")
print("="*60)
df_model = df.copy()
df_model['IsNewCustomer'] = (df_model['tenure'] < 12).astype(int)
print(f"Created: IsNewCustomer")
service_cols = ['PhoneService', 'InternetService', 'OnlineSecurity', 
                'OnlineBackup', 'DeviceProtection', 'TechSupport', 
                'StreamingTV', 'StreamingMovies']

def count_services(row):
    count = 0
    for col in service_cols:
        if row[col] not in ['No', 'No phone service', 'No internet service']:
            count += 1
    return count

df_model['ServiceCount'] = df_model.apply(count_services, axis=1)
print(f"Created: ServiceCount")


df_model['HighRiskCombo'] = (
    (df_model['MonthlyCharges'] > 70) & 
    (df_model['tenure'] < 12)
).astype(int)
print(f"Created: HighRiskCombo")

print("\nNew features preview:")
print(df_model[['tenure', 'IsNewCustomer', 'ServiceCount', 
                'MonthlyCharges', 'HighRiskCombo', 'Churn']].head(10))

print(f"\nTotal features now: {df_model.shape[1]}")
```

```python
# Cell 10: Data Preprocessing

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

print("="*60)
print("DATA PREPROCESSING")
print("="*60)
y = df_model['Churn'].map({'Yes': 1, 'No': 0})
print(f"Target variable (y) created: {y.shape}")

cols_to_drop = ['customerID', 'Churn']
X = df_model.drop(cols_to_drop, axis=1)
print(f"Features (X) created: {X.shape}")

binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
for col in binary_cols:
    X[col] = X[col].map({'Yes': 1, 'No': 0})
print(f"Converted {len(binary_cols)} binary columns")

for col in X.columns:
    if X[col].dtype == 'object':
        X[col] = X[col].replace(['No internet service', 'No phone service'], 'No')

# get remaining categorical columns
cat_cols = X.select_dtypes(include=['object']).columns.tolist()
print(f"Categorical columns to encode: {cat_cols}")

# One-hot encoding
X = pd.get_dummies(X, columns=cat_cols, drop_first=True)
print(f"One-hot encoding complete")
print(f"Final feature count: {X.shape[1]}")

# Train-test split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% for testing
    random_state=42,    # For reproducibility
    stratify=y          # Keep same churn ratio in both sets
)

print(f"\nTrain-Test Split Complete")
print(f"Training set: {X_train.shape[0]} customers")
print(f"Test set: {X_test.shape[0]} customers")
print(f"Train churn rate: {y_train.mean():.2%}")
print(f"Test churn rate: {y_test.mean():.2%}")

# Scale numerical features
num_cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

scaler = StandardScaler()
X_train[num_cols_to_scale] = scaler.fit_transform(X_train[num_cols_to_scale])
X_test[num_cols_to_scale] = scaler.transform(X_test[num_cols_to_scale])

print(f"\nScaled {len(num_cols_to_scale)} numerical columns")
print("\n" + "="*60)
print("READY FOR MODELING!")
print("="*60)
```

```python
#Baseline Model - Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve

print("="*60)
print("MODEL 1: LOGISTIC REGRESSION (BASELINE)")
print("="*60)

# Train model
lr_model = LogisticRegression(
    random_state=42,
    max_iter=1000,
    class_weight='balanced'  
)

print("Training model...")
lr_model.fit(X_train, y_train)
print("Training complete!")

# Make predictions
y_pred_lr = lr_model.predict(X_test)
y_pred_proba_lr = lr_model.predict_proba(X_test)[:, 1]

# Evaluate
auc_lr = roc_auc_score(y_test, y_pred_proba_lr)

print(f"\n{'='*60}")
print(f"ROC-AUC Score: {auc_lr:.3f}")
print(f"{'='*60}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr, 
                          target_names=['No Churn', 'Churn'],
                          digits=3))

cm_lr = confusion_matrix(y_test, y_pred_lr)
print("Confusion Matrix:")
print(cm_lr)
print(f"\nInterpretation:")
print(f"- Correctly predicted No Churn: {cm_lr[0,0]}")
print(f"- Incorrectly predicted Churn (False Alarm): {cm_lr[0,1]}")
print(f"- Missed Churners (Worst!): {cm_lr[1,0]}")
print(f"- Correctly predicted Churn: {cm_lr[1,1]}")

# Visualize Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Churn', 'Churn'],
            yticklabels=['No Churn', 'Churn'],
            cbar_kws={'label': 'Count'})
plt.ylabel('Actual', fontsize=12, fontweight='bold')
plt.xlabel('Predicted', fontsize=12, fontweight='bold')
plt.title('Confusion Matrix - Logistic Regression', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('../images/05_confusion_matrix_lr.png', dpi=150, bbox_inches='tight')
plt.show()

# ROC Curve
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_pred_proba_lr)

plt.figure(figsize=(8, 6))
plt.plot(fpr_lr, tpr_lr, linewidth=2, 
         label=f'Logistic Regression (AUC = {auc_lr:.3f})')
plt.plot([0, 1], [0, 1], 'k--', linewidth=2, label='Random Guess')
plt.xlabel('False Positive Rate', fontsize=12, fontweight='bold')
plt.ylabel('True Positive Rate (Recall)', fontsize=12, fontweight='bold')
plt.title('ROC Curve - Logistic Regression', fontsize=14, fontweight='bold', pad=20)
plt.legend(fontsize=11)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('../images/06_roc_curve_lr.png', dpi=150, bbox_inches='tight')
plt.show()


print("\nBaseline model complete!")
print(f"Plots saved to images/")
```

```python
# Cell 12: Advanced Model - XGBoost

from xgboost import XGBClassifier

print("="*60)
print("MODEL 2: XGBOOST (ADVANCED)")
print("="*60)

scale_pos_weight = len(y_train[y_train==0]) / len(y_train[y_train==1])
print(f"Class imbalance ratio: {scale_pos_weight:.2f}")

# Train XGBoost
xgb_model = XGBClassifier(
    n_estimators=200,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='auc',
    scale_pos_weight=scale_pos_weight
)

print("\nTraining XGBoost model...")
xgb_model.fit(X_train, y_train, verbose=False)
print("Training complete!")

# Predictions
y_pred_xgb = xgb_model.predict(X_test)
y_pred_proba_xgb = xgb_model.predict_proba(X_test)[:, 1]

# Evaluate
auc_xgb = roc_auc_score(y_test, y_pred_proba_xgb)

print(f"\n{'='*60}")
print(f"ROC-AUC Score: {auc_xgb:.3f}")
print(f"{'='*60}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_xgb,
                          target_names=['No Churn', 'Churn'],
                          digits=3))

# Confusion Matrix
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
print("Confusion Matrix:")
print(cm_xgb)

# Visualize
plt.figure(figsize=(8, 6))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Greens',
            xticklabels=['No Churn', 'Churn'],
            yticklabels=['No Churn', 'Churn'],
            cbar_kws={'label': 'Count'})
plt.ylabel('Actual', fontsize=12, fontweight='bold')
plt.xlabel('Predicted', fontsize=12, fontweight='bold')
plt.title('Confusion Matrix - XGBoost', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('../images/07_confusion_matrix_xgb.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nXGBoost model complete!")
```

```python

print("="*60)
print("MODEL COMPARISON")
print("="*60)

# Create comparison table
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'XGBoost'],
    'ROC-AUC': [auc_lr, auc_xgb],
    'Precision': [
        classification_report(y_test, y_pred_lr, output_dict=True)['1']['precision'],
        classification_report(y_test, y_pred_xgb, output_dict=True)['1']['precision']
    ],
    'Recall': [
        classification_report(y_test, y_pred_lr, output_dict=True)['1']['recall'],
        classification_report(y_test, y_pred_xgb, output_dict=True)['1']['recall']
    ],
    'F1-Score': [
        classification_report(y_test, y_pred_lr, output_dict=True)['1']['f1-score'],
        classification_report(y_test, y_pred_xgb, output_dict=True)['1']['f1-score']
    ]
}).round(3)

print(results.to_string(index=False))

# Visualize comparison
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# ROC Curves
axes[0].plot(fpr_lr, tpr_lr, linewidth=2, label=f'Logistic Reg (AUC={auc_lr:.3f})')
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_pred_proba_xgb)
axes[0].plot(fpr_xgb, tpr_xgb, linewidth=2, label=f'XGBoost (AUC={auc_xgb:.3f})')
axes[0].plot([0, 1], [0, 1], 'k--', alpha=0.3)
axes[0].set_xlabel('False Positive Rate', fontsize=11, fontweight='bold')
axes[0].set_ylabel('True Positive Rate', fontsize=11, fontweight='bold')
axes[0].set_title('ROC Curves Comparison', fontsize=13, fontweight='bold')
axes[0].legend(fontsize=10)
axes[0].grid(alpha=0.3)

# Metrics comparison
metrics_df = results.set_index('Model')[['Precision', 'Recall', 'F1-Score']]
metrics_df.plot(kind='bar', ax=axes[1], width=0.8)
axes[1].set_ylabel('Score', fontsize=11, fontweight='bold')
axes[1].set_title('Performance Metrics Comparison', fontsize=13, fontweight='bold')
axes[1].set_ylim(0.5, 0.9)
axes[1].legend(fontsize=10)
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45)
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../images/08_model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

# Determine winner
if auc_xgb > auc_lr:
    winner = "XGBoost"
    best_model = xgb_model
    best_proba = y_pred_proba_xgb
    best_auc = auc_xgb
else:
    winner = "Logistic Regression"
    best_model = lr_model
    best_proba = y_pred_proba_lr
    best_auc = auc_lr

print(f"\n{'='*60}")
print(f"WINNER: {winner} with {best_auc:.3f} AUC")
print(f"{'='*60}")
print("\nModel comparison complete!")
```

```python

print("="*60)
print("FEATURE IMPORTANCE - What Drives Churn?")
print("="*60)

feature_importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': xgb_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nTop 15 Most Important Features:")
print(feature_importance_df.head(15).to_string(index=False))

# Visualize top 15
plt.figure(figsize=(10, 8))
top_15 = feature_importance_df.head(15)
colors = plt.cm.viridis(np.linspace(0.3, 0.9, 15))

plt.barh(range(len(top_15)), top_15['Importance'], color=colors)
plt.yticks(range(len(top_15)), top_15['Feature'])
plt.xlabel('Importance Score', fontsize=12, fontweight='bold')
plt.title('Top 15 Features for Churn Prediction', fontsize=14, fontweight='bold', pad=20)
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('../images/09_feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nFeature importance plot saved!")
```

```python
import shap
print("="*60)
print("SHAP ANALYSIS - Why Does the Model Predict Churn?")
print("="*60)

# Sample 500 customers for faster computation
X_test_sample = X_test.sample(min(500, len(X_test)), random_state=42)

print(f"Computing SHAP values for {len(X_test_sample)} customers...")
print("(This may take 1-2 minutes...)")

# Create SHAP explainer
explainer = shap.TreeExplainer(xgb_model)
shap_values = explainer.shap_values(X_test_sample)

print("SHAP computation complete!")

# SHAP Summary Plot - Feature Importance
print("\nGenerating SHAP summary plots...")

plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values, X_test_sample, plot_type="bar", show=False)
plt.title('SHAP Feature Importance', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('../images/10_shap_importance.png', dpi=150, bbox_inches='tight')
plt.show()
plt.figure(figsize=(10, 8))
shap.summary_plot(shap_values, X_test_sample, show=False)
plt.tight_layout()
plt.savefig('../images/11_shap_detailed.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nSHAP plots saved!")
print("\n" + "="*60)
```

```python
# Cell 16: Business Impact & ROI Analysis

print("="*60)
print("BUSINESS IMPACT ANALYSIS")
print("="*60)

# Business assumptions (you'd get these from stakeholders)
CUSTOMER_LIFETIME_VALUE = 2000  # Average revenue from retained customer
RETENTION_COST = 100  # Cost of retention offer per customer
RETENTION_SUCCESS_RATE = 0.30  # 30% of interventions work
results_df = pd.DataFrame({
    'ChurnProbability': best_proba,
    'ActualChurn': y_test.values
}, index=y_test.index)
results_df['RiskTier'] = pd.cut(
    results_df['ChurnProbability'],
    bins=[0, 0.3, 0.7, 1.0],
    labels=['Low Risk', 'Medium Risk', 'High Risk']
)

print("\nCustomer Risk Segmentation:")
print(results_df['RiskTier'].value_counts().sort_index())

# Focus on HIGH RISK customers
high_risk_customers = results_df[results_df['RiskTier'] == 'High Risk']

print(f"\n{'='*60}")
print("HIGH RISK CUSTOMERS ANALYSIS")
print(f"{'='*60}")

n_high_risk = len(high_risk_customers)
actual_churners_in_high_risk = high_risk_customers['ActualChurn'].sum()
catch_rate = (actual_churners_in_high_risk / y_test.sum()) * 100

print(f"Number of high-risk customers: {n_high_risk}")
print(f"Actual churners in this group: {actual_churners_in_high_risk}")
print(f"% of all churners we identified: {catch_rate:.1f}%")

# ROI Calculation
intervention_cost = n_high_risk * RETENTION_COST
saved_customers = actual_churners_in_high_risk * RETENTION_SUCCESS_RATE
revenue_saved = saved_customers * CUSTOMER_LIFETIME_VALUE
net_benefit = revenue_saved - intervention_cost
roi = (net_benefit / intervention_cost) * 100

print(f"\n{'='*60}")
print("ROI CALCULATION")
print(f"{'='*60}")
print(f"Customers to target: {n_high_risk}")
print(f"Intervention cost: ${intervention_cost:,.0f}")
print(f"Expected customers saved: {saved_customers:.0f}")
print(f"Revenue saved: ${revenue_saved:,.0f}")
print(f"Net benefit: ${net_benefit:,.0f}")
print(f"ROI: {roi:.1f}%")
print(f"{'='*60}")

# Visualize Risk Distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Risk tier distribution
risk_counts = results_df['RiskTier'].value_counts().sort_index()
colors_risk = ['#2ecc71', '#f39c12', '#e74c3c']
axes[0].bar(range(len(risk_counts)), risk_counts.values, color=colors_risk)
axes[0].set_xticks(range(len(risk_counts)))
axes[0].set_xticklabels(risk_counts.index, rotation=0)
axes[0].set_ylabel('Number of Customers', fontsize=11, fontweight='bold')
axes[0].set_title('Customer Distribution by Risk Tier', fontsize=13, fontweight='bold')
axes[0].grid(axis='y', alpha=0.3)

for i, v in enumerate(risk_counts.values):
    axes[0].text(i, v + 10, str(v), ha='center', va='bottom', fontweight='bold')

# Churn rate by risk tier
churn_by_risk = results_df.groupby('RiskTier')['ActualChurn'].apply(
    lambda x: x.sum() / len(x) * 100
).sort_index()
axes[1].bar(range(len(churn_by_risk)), churn_by_risk.values, color=colors_risk)
axes[1].set_xticks(range(len(churn_by_risk)))
axes[1].set_xticklabels(churn_by_risk.index, rotation=0)
axes[1].set_ylabel('Actual Churn Rate (%)', fontsize=11, fontweight='bold')
axes[1].set_title('Actual Churn Rate by Risk Tier', fontsize=13, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

for i, v in enumerate(churn_by_risk.values):
    axes[1].text(i, v + 2, f'{v:.1f}%', ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.savefig('../images/12_business_impact.png', dpi=150, bbox_inches='tight')
plt.show()
print("\nBusiness impact analysis complete!")
```
